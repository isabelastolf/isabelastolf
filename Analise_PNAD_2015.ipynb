# Checking Gender, Color and State
<p> Gender (Sexo), Color (Cor) and State (UF)

```python
# as the database talks about the family chief information, I wanted to 
# understand better the dynamics
```

```python
# are there more male or female chiefs?
data['Sexo'].value_counts()
```

```python
gender_frequency = data['Sexo'].value_counts()
gender_percentage = data['Sexo'].value_counts(normalize = True) * 100
freq_gender = pd.DataFrame({'Frequency': gender_frequency, 'Percentage (%)': gender_percentage})
freq_gender.rename(index = {0: 'Male', 1: 'Female'}, inplace = True)
freq_gender.rename_axis('Gender', axis = 'columns', inplace = True)
```

```python
freq_gender
```

```python
# which color is more frequent between chiefs
color_frequency = data['Cor'].value_counts()
color_percentage = data['Cor'].value_counts(normalize = True) * 100
freq_color = pd.DataFrame({'Frequency': color_frequency, 'Percentage (%)': color_percentage})
freq_color.rename(index = {0: 'Indigenous', 2: 'White', 4: 'Black', 6: 'Yellow', 8: 'Brown', 9: 'Not declared'}, inplace = True)
freq_color.rename_axis('Color', axis = 'columns', inplace = True)
```

```python
freq_color
```

```python
# which UF has the most family chiefs
UF_frequency = data['UF'].value_counts()
UF_percentage = data['UF'].value_counts(normalize = True) * 100
freq_UF = pd.DataFrame({'Frequency': UF_frequency, 'Percentage (%)': UF_percentage})
freq_UF.rename(index = {11:	'Rondônia', 12:	'Acre', 13:	'Amazonas', 14:	'Roraima', 15:	'Pará', 16:	'Amapá', 17:	'Tocantins', 
                        21:	'Maranhão', 22:	'Piauí', 23:	'Ceará', 24:	'Rio Grande do Norte', 25:	'Paraíba', 26:	'Pernambuco', 27:	'Alagoas', 28:	'Sergipe', 29:	'Bahia',
                        31:	'Minas Gerais', 32:	'Espírito Santo', 33:	'Rio de Janeiro', 35:	'São Paulo', 
                        41:	'Paraná', 42:	'Santa Catarina', 43:	'Rio Grande do Sul', 
                        50:	'Mato Grosso do Sul', 51:	'Mato Grosso', 52:	'Goiás', 53:	'Distrito Federal'}, inplace = True)
freq_UF.rename_axis('UF', axis = 'columns', inplace = True)
```

# Checking the Income (Renda)

```python
data.Renda.min()
```

```python
data.Renda.max()
```

```python
classes = [0, 1576, 3152, 7880, 15760, 200000]
```

```python
labels = ['E', 'D', 'C', 'B', 'A']
```

```python
pd.cut(x = data.Renda,
       bins = classes,
       labels = labels,
       include_lowest = True)
```

```python
frequency_income = pd.value_counts(
    pd.cut(x = data.Renda,
       bins = classes,
       labels = labels,
       include_lowest = True)
)
frequency_income
```

```python
percentage_income = pd.value_counts(
    pd.cut(x = data.Renda,
       bins = classes,
       labels = labels,
       include_lowest = True),
       normalize = True
) * 100
percentage_income = percentage_income.round()
percentage_income
```

```python
freq_income = pd.DataFrame({'Frequency': frequency_income, 'Percentage (%)': percentage_income})
freq_income
freq_income.sort_index(ascending = False)
```
